```python
import pandas as pd
print(pd.__version__)
```

```python
auto_data = pd.read_csv('./data/imports-85.data',sep=r'\s*,\s*',engine='python')
auto_data
```

<p>missing datas are represented with question mark.Replacing ? with nan using numpy lib <p>

```python
import numpy as np
auto_data = auto_data.replace('?',np.nan)
auto_data.head()
```

```python
auto_data.describe(include='all')
```

```python
auto_data['price'].describe()
```

<p>price column look like object..its should convert to float.using pandas to_numeric function for this conversion</p>

```python
auto_data['price'] = pd.to_numeric(auto_data['price'], errors='coerce')
auto_data['horsepower'] = pd.to_numeric(auto_data['horsepower'], errors='coerce')
auto_data['price'].describe()
```

<p>now remove columns not related to price</p> 

```python
auto_data = auto_data.drop('normalized-losses',axis=1)
```

```python
auto_data.head()
```

```python
auto_data['num-of-cylinders'].describe()
```

```python
cylinders_dict = {'two': 2,
                  'three': 3,
                  'four':4,
                  'five': 5,
                  'six':6,
                  'eight':8,
                  'twelve':12
                 }
auto_data['num-of-cylinders'].replace(cylinders_dict, inplace = True)
auto_data.head()
```

<p>convert data in to one hot representation</p>

```python
auto_data = pd.get_dummies(auto_data,columns=['make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','fuel-system'])
```

```python
auto_data.head()
```

<p>cleaned up the data :: drop the row that have not a number value</p>

```python
auto_data = auto_data.dropna()
auto_data
```

<p>checking data set have any null value</p>

```python
auto_data[auto_data.isnull().any(axis=1)]
```

<p>now data is good to go for michine learning model.taiking datasets 80% for training and 20% for testing</p>

```python
from sklearn.model_selection import train_test_split

x = auto_data.drop('price', axis =1)

y = auto_data['price']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
```

```python
linear_model.score(x_train,y_train)
```

our R-square value for the prediction is : 0.9679

```python
linear_model.coef_
```

will display the coefficiency of each value..the weights of all featurs in our data set

```python
predictors = x_train.columns
coef = pd.Series(linear_model.coef_,predictors).sort_values()
print(coef)
```

above showing indvitual coefficiant featurs of the data.
it will help to understand what and all factors are affecting the automobile price.
-ve weightage means "car with those featurs are cheap"


```python
y_predict = linear_model.predict(x_test)
```
